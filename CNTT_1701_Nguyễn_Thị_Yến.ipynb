```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc file

df = pd.read_excel("/content/drive/MyDrive/CNTT17-01_Nguyễn _Thị_Yến_BKTS2(1).xlsx") #(YẾN)

giá_trị_thay_thế = 0;

# Xử lí giá trị bị thiếu
df["Giá (VNĐ)"].fillna(df["Giá (VNĐ)"].median(), inplace=True) #Dùng để điền giá trị còn thiếu trong cột Giá bằng giá trị trung vị (là giá trị đứng ở giữa khi sắp xếp các số theo thứ tự tăng dần)
df["Số Lượng"].fillna(giá_trị_thay_thế, inplace=True)
df["Ngày Nhập"].fillna("NaN", inplace = True) # Đánh dấu ngày nhập thiếu
df["Bảo Hành (tháng)"].fillna(df["Bảo Hành (tháng)"].median(), inplace = True) #Lấy giá trị trung vị của cột "Bảo hành (Tháng)".

# Sửa dữ liệu sai lệch
df["Số Lượng"] = df["Số Lượng"].apply(lambda x: abs(x) if x < 0 else x)  # Chuyển số lượng âm thành dương
df["Bảo Hành (tháng)"] = df["Bảo Hành (tháng)"].apply(lambda x: abs(x) if x < 0 else x)  # Chuyển bảo hành âm thành dương

# Xuất file Excel đã làm sạch
df.to_excel("/content/drive/MyDrive/CNTT17-01_Nguyễn _Thị_Yến_BKTS2(1).xlsx", index=False)

print("Dữ liệu đã được làm sạch và lưu vào file mới.")

#khám phá dữ liệu
print("Thông tin tổng quan về dữ liệu:")
print(df.info())
print("\nThống kê mô tả:")
print(df.describe())

# Biểu đồ phân bố bảo hành sản phẩm
plt.figure(figsize=(8, 5))
sns.histplot(df["Bảo Hành (tháng)"], bins=15, kde=True)
plt.title("Phân phối thời gian bảo hành")
plt.xlabel("Bảo Hành (tháng)")
plt.ylabel("Tần suất")
plt.show()
```

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Chuyển dữ liệu thành DataFrame
df = pd.DataFrame(df)

# Chuyển đổi thành mảng numpy
X = df["Số Lượng"].values.reshape(-1, 1)  # Biến độc lập (reshape thành ma trận cột)
y = df["Giá (VNĐ)"].values  # Biến phụ thuộc

# 5️ Xây dựng mô hình hồi quy tuyến tính
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

#  Tính toán hệ số hồi quy
beta_0 = model.intercept_  # Hệ số chặn
beta_1 = model.coef_[0]    # Hệ số góc

# Đánh giá mô hình
r2 = r2_score(y, y_pred)  # Hệ số xác định R^2
mse = mean_squared_error(y, y_pred)  # Sai số bình phương trung bình (MSE)

#  Hiển thị kết quả
print("\n🔹 Kết quả hồi quy tuyến tính:")
print(f"Phương trình hồi quy: y = {beta_0:.2f} + {beta_1:.2f}x")
print(f"Hệ số chặn (Intercept - beta_0): {beta_0}")
print(f"Hệ số góc (Slope - beta_1): {beta_1}")
print(f"Hệ số xác định (R²): {r2:.4f}")
print(f"Sai số bình phương trung bình (MSE): {mse:.4f}")

#  Vẽ biểu đồ
plt.scatter(X, y, color="blue", label="Dữ liệu thực tế")
plt.plot(X, y_pred, color="red", label="Hồi quy tuyến tính")
plt.title("Hồi quy tuyến tính: Số Lượng và Giá")
plt.xlabel("Số Lượng")
plt.ylabel("Giá")
plt.legend()
plt.grid(True)
plt.show()
```

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```
